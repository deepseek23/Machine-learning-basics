## K Means Clustering


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_moons
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler 
```

```python
X, y_true = make_moons(n_samples=500, noise=0.1, random_state=42)
```

```python
df = pd.DataFrame(X, columns=["Feature_1","Feature_2"])
```

```python
df
```

```python
scaler = StandardScaler()
```

```python
X_scaled = scaler.fit_transform(df)
```

```python
intertia = []
k_range = range(1,11)
```

```python
for k in k_range:
    model = KMeans(n_clusters=k, random_state=42)
    model.fit(X_scaled)
    intertia.append(model.inertia_)
```

```python
intertia
```

```python
plt.plot(k_range, intertia, marker='o')
```

```python
kmeans_final = KMeans(n_clusters=2, random_state=42)
```

```python
cluster_labels = kmeans_final.fit_predict(X_scaled)
```

```python
cluster_labels
```

```python
df['cluster'] = cluster_labels
```

```python
sns.scatterplot(
    x = df['Feature_1'],
    y = df['Feature_2'],
    hue=df['cluster'],
    palette='viridis'
)
```

## DB (Density Base) Scan Algorithm

```python
from sklearn.datasets import make_moons
```

```python
X, y_true = make_moons(n_samples=500, noise=0.05, random_state=42)
```

```python
X
```

```python
from sklearn.cluster import KMeans, DBSCAN
```

```python
dff =  pd.DataFrame(X, columns=["Feature_1","Feature_2"])
```

```python
sns.scatterplot(x=dff['Feature_1'], y=dff['Feature_2'])
```

```python
x_n =  scaler.fit_transform(dff)
```

```python
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans_labels = kmeans.fit_predict(x_n)
```

```python

dff['kmeans_cluster'] = kmeans_labels
```

```python

sns.scatterplot(x=dff['Feature_1'], y=dff['Feature_2'], hue=dff['kmeans_cluster'], palette='tab10')
```

```python
dbscan = DBSCAN(eps=0.3, min_samples=5)
dbscan_labels = dbscan.fit_predict(x_n)
dff['dbscan_clusters'] = dbscan_labels



```

```python
 
sns.scatterplot(x=dff['Feature_1'], y=dff['Feature_2'], hue=dff['dbscan_clusters'], palette='tab10')
```
