```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df =  sns.load_dataset('iris')
```

```python
df.head()
```

```python
df['species'].unique()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('species', axis=1)
y = df['species']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model_knn = KNeighborsClassifier(n_neighbors=5)
```

```python
model_knn.fit(X_train, y_train)
```

```python
model_knn.score(X_test,y_test)
```

```python
from sklearn.svm import SVC
```

```python
model_svc = SVC(gamma= 'auto')
```

```python
model_svc.fit(X_train, y_train)
```

```python
model_svc.score(X_test,y_test)
```

# now lets use grid search cv

```python
from sklearn.model_selection import GridSearchCV
```

```python
classifier = GridSearchCV((model_svc),{
    'C': [1,10,20,30],
    'kernel': ['rbf','linear'],

},cv=5,return_train_score=False)
```

```python
classifier.fit(X,y)
```

```python
classifier.cv_results_
```

```python
results = pd.DataFrame(classifier.cv_results_)
```

```python
results
```

```python

results[['param_C','param_kernel','mean_test_score']]
```

```python
classifier_knn = GridSearchCV((model_knn),{
    'n_neighbors': [3,5,7,9,11,13],
    'weights': ['uniform','distance'],
    'metric': ['euclidean', 'manhattan', 'minkowski'],

},cv=5,return_train_score=False)
```

```python
classifier_knn.fit(X,y)
```

```python
results_knn = pd.DataFrame(classifier_knn.cv_results_)
```

```python
# Rank KNN configs by accuracy (higher is better)
results_knn_sorted = results_knn[['param_n_neighbors', 'param_weights', 'mean_test_score']].copy()
results_knn_sorted['rank'] = results_knn_sorted['mean_test_score'].rank(ascending=False, method='dense').astype(int)
results_knn_sorted.sort_values('mean_test_score', ascending=False).reset_index(drop=True)
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
classifier_r = RandomizedSearchCV((model_svc),{
    'C': [1,10,20,30],
    'kernel': ['rbf','linear'],

},n_iter=203 ,cv=5,return_train_score=False)
```

```python
classifier_r.fit(X,y)
```

```python
r_svm = pd.DataFrame(classifier_r.cv_results_)
```
